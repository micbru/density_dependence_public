Compare $\Pi$ bisection predictions from METE, random placement, and density dependent death model.

### METE
$$
    \Pi_{\text{METE}}(n) = \frac{1}{n_0+1}
$$

### Random Placement
$$
    \Pi_{\text{RP}}(n) = \binom{n_0}{n} \left(\frac{1}{2}\right)^{n_0}
$$

### Density dependent death with general $\alpha$
$$
    \Pi_{\alpha}(n) = \frac{n^\alpha + (n_0-n)^\alpha}{C n_0^\alpha} \binom{n_0}{n}^{\alpha-1}
$$

### Density dependent death with $\alpha=2$
$$
    \Pi_{\alpha=2}(n) = \frac{n^2 + (n_0-n)^2}{2^{n_0-1} n_0 (n_0+1)} \binom{n_0}{n}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import dd_functions as dd
%matplotlib inline

import scipy.special
```

```python
# Define other functions

def METE(n,n0):
    return 1/(n0+1)

def RP(n,n0):
    return scipy.special.binom(n0, n)*(0.5)**n0
```

# Small n0

```python
n0=10
n = np.arange(n0+1)
```

```python
plt.figure(figsize=(4,3))
plt.plot(n,(n0+1)*[METE(n,n0)],'-',linewidth=2,label='METE')
plt.plot(n,RP(n,n0),'-',linewidth=2,label='RP')
plt.plot(n,dd.dd_prob(n0,1.1),'--',label=r'$\alpha=1.1$')
plt.plot(n,dd.dd_prob(n0,1.5),'-.',label=r'$\alpha=1.5$')
plt.plot(n,dd.dd_prob(n0,1.9),':',label=r'$\alpha=1.9$')
plt.legend()
plt.ylim(0,0.25)
plt.xlim(0,10)
plt.xlabel(r'$n$')
plt.ylabel(r'$\Pi(n)$')
plt.title('$n_0$=10')
fig_name = 'FigureFilename.pdf'
plt.savefig(fig_name,format='pdf', bbox_inches='tight')
plt.show()
```
